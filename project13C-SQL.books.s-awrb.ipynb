### Computing the number of user reviews and the average rating for each book
    In this session, for computing the number of user reviews and the average rating for each book, we should some steps below:

    1. We are going to use CTE (Common Table Expression) for creating two new tables of:
        a. Getting the reviews number of each book that we will create as table number one.
        b. Getting the ratings average of each book that we will create as table number two.

    2. Both of new tables are going to merge by using INNER JOIN method at 'book_id'.
    
    The execution of computing number of user reviews and average rating for each book, as follows:

    1. CTE (Common Table Expression)

```python
pd.io.sql.read_sql(
'''
WITH agg_reviews_per_book AS (
    SELECT
        reviews.book_id,
        COUNT(reviews.text) AS cnt_reviews,
        books.title
    FROM
        reviews
        INNER JOIN books ON books.book_id = reviews.book_id
    GROUP BY
        reviews.book_id,
        books.title
    ORDER BY
        cnt_reviews DESC
)
, avg_ratings_per_book AS(
    SELECT
        ratings.book_id,
        AVG(ratings.rating) AS avg_ratings
    FROM 
        ratings
    GROUP BY
        ratings.book_id
    ORDER BY
        avg_ratings DESC
)
SELECT 
    agg_reviews_per_book.title,
    cnt_reviews,
    avg_ratings
FROM
    agg_reviews_per_book
    INNER JOIN avg_ratings_per_book on agg_reviews_per_book.book_id = avg_ratings_per_book.book_id
''',
con = engine)
```

    2. INNER JOIN method at 'book_id'

```python
pd.io.sql.read_sql(
'''
WITH main AS (
    WITH agg_reviews_per_book AS (
        SELECT
            reviews.book_id,
            COUNT(reviews.text) AS cnt_reviews,
            books.title
        FROM
            reviews
            INNER JOIN books ON books.book_id = reviews.book_id
        GROUP BY
            reviews.book_id,
            books.title
        ORDER BY
            cnt_reviews DESC
    )
    , avg_ratings_per_book AS(
        SELECT
            ratings.book_id,
            AVG(ratings.rating) AS avg_ratings
        FROM 
            ratings
        GROUP BY
            ratings.book_id
        ORDER BY
            avg_ratings DESC
    )
    SELECT 
        agg_reviews_per_book.title,
        cnt_reviews,
        avg_ratings
    FROM
        agg_reviews_per_book
        INNER JOIN avg_ratings_per_book on agg_reviews_per_book.book_id = avg_ratings_per_book.book_id
    )
SELECT
    MIN(cnt_reviews) AS min_reviews_per_book,
    MAX(cnt_reviews) AS max_reviews_per_book,
    MIN(avg_ratings) AS min_ratings_per_book,
    MAX(avg_ratings) AS max_ratings_per_book
FROM
    main
''',
con = engine)
```

    Findings:
    
    - From the tables above, the books that receive reviews from users are 994 books.
    - The range of reviews per book is 1 of 7 reviews.
    - The range of ratings average is from scale 1.5 to 5.0.

### Identifying the publisher that has published the most number of books, with more than 50 pages
    - This will assist us in excluding brochures and similar publications from the analysis.

    We are going to identify the publisher that has published the most number of books, with more than 50 pages, as follows:

```python
pd.io.sql.read_sql(
'''
SELECT 
    publisher,
    COUNT(publisher) AS cnt
FROM 
    publishers
    LEFT JOIN books ON books.publisher_id = publishers.publisher_id
WHERE
    books.num_pages > 50
GROUP BY
    publisher
ORDER BY
    cnt DESC;
''',
con = engine)
```

    Findings:
    
    - Based on data above, there are 334 publishers that has published the most number of books with more than 50 pages.

### Identifying authors with the highest average book rating
    - Finding books with the rating of at least 50

    First, we are going to identify the author with the highest average book rating, as follows:

```python
pd.io.sql.read_sql(
'''
SELECT
    authors.author_id,
    authors.author,
    COUNT(ratings.rating) AS cnt_rating,
    ROUND(AVG(ratings.rating), 2) AS avg_rating
FROM 
    books
    LEFT JOIN authors
        ON books.author_id = authors.author_id
    LEFT JOIN ratings
        ON books.book_id = ratings.book_id
GROUP BY
    authors.author_id,
    authors.author
HAVING
    COUNT(ratings.rating) >= 50
ORDER BY
    avg_rating DESC;
''',
con = engine)
```

    Findings:
    
    - There are 24 authors with the highest average book rating.

### Computing the average number of review texts among users who have rated more than 50 books
    In this session, we will do as the same previously method, those are:

    1. We are going to use CTE (Common Table Expression) for creating two new tables of:
    a. Getting the number of users who rated more than 50 books that we will create as table one.
    b. Getting the number of users who review books that we will create as table two.

    2. Both of new tables are going to merge by using INNER JOIN method at 'username'.

    The execution of computing average number of review texts among users who have rated more than 50 books, as follows:

    CTE (Common Table Expression) and INNER JOIN method at 'username':

```python
pd.io.sql.read_sql(
'''
WITH user_more_than_50_ratings AS (
    SELECT
        username,
        COUNT(username) AS cnt_rating
    FROM 
        ratings
    GROUP BY
        username
    HAVING
        COUNT(username) > 50
)
, main AS (
    SELECT
        reviews.username,
        COUNT(reviews.review_id) AS cnt_reviews
    FROM
        reviews
        INNER JOIN user_more_than_50_ratings ON reviews.username = user_more_than_50_ratings.username
    GROUP BY
        reviews.username
    ORDER BY
        cnt_reviews DESC
)
SELECT
    ROUND(AVG(cnt_reviews), 0) AS avg_reviews
FROM
    main
''',
con = engine)
```

    Findings:
    
    - Based on the result above, the average number of review texts among users who have rated more than 50 books is 24.0 (24 reviews).
