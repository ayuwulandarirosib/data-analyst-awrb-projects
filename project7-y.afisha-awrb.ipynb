#### Category Data Type

    We will analyze the column one by one by taking 'Device' column. Then, we'll find the number of unique values in this column and see how they are distributed. We'll use the .value_counts() method:

```python
visits_data['Device'].value_counts()
```

    Findings:

    - The data above, this column only has two values, namely 'desktop' and 'touch'. These two values represent the categories of devices ('devices') that users use: computers ('desktops') and touch screen devices ('touch').

    - Pandas has a special data type for such data, namely 'category'. This data type looks very similar to the 'object' type, except that it takes up much less memory space. To convert data to type 'category', we will use the astype('category') method:

```python
visits_data['Device'] = visits_data['Device'].astype('category')
```

```python
visits_data.head()
```

```python
visits_data.info(memory_usage='deep')
```

    - The Dataframe size has changed, now it is 57.9 MB instead of 79.3 MB. We've managed to make it smaller by changing the data type of one of the columns. Next, we'll see what other things we could do.
